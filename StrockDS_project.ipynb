```python

def count_outliers(column):
    Q1 = column.quantile(0.25)  # First quartile
    Q3 = column.quantile(0.75)  # Third quartile
    IQR = Q3 - Q1               # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Count outliers
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return len(outliers)

# Check outliers for each column
print("Outliers in bmi:", count_outliers(df['bmi']))
print("Outliers in avg_glucose_level:", count_outliers(df['avg_glucose_level']))
print("Outliers in age:", count_outliers(df['age']))

```

```python
import numpy as np
# Calculate IQR bounds for bmi
Q1_bmi = df['bmi'].quantile(0.25)
Q3_bmi = df['bmi'].quantile(0.75)
IQR_bmi = Q3_bmi - Q1_bmi
lol = Q1_bmi - 1.5 * IQR_bmi  # Lower bound
upl = Q3_bmi + 1.5 * IQR_bmi  # Upper bound

# Calculate IQR bounds for avg_glucose_level
Q1_glucose = df['avg_glucose_level'].quantile(0.25)
Q3_glucose = df['avg_glucose_level'].quantile(0.75)
IQR_glucose = Q3_glucose - Q1_glucose
ll = Q1_glucose - 1.5 * IQR_glucose  # Lower bound
ul = Q3_glucose + 1.5 * IQR_glucose  # Upper bound


df['bmi']=np.where(df['bmi']>upl,upl,np.where(df['bmi']<lol,lol,df['bmi']))
df['avg_glucose_level']=np.where(df['avg_glucose_level']>ul,ul,np.where(df['avg_glucose_level']<ll,ll,df['avg_glucose_level']))

n=len(df.loc[~df['bmi'].between(lol,upl)])
print("The number of outliers in bmi is: ",n)


nn=len(df.loc[~df['avg_glucose_level'].between(ll,ul)])
print("The number of outliers in avg_glucose_level is: ",nn)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['age', 'avg_glucose_level', 'bmi']] = scaler.fit_transform(df[['age', 'avg_glucose_level', 'bmi']])
```

```python
# Label encoding for binary categorical variables
df['gender'] = df['gender'].astype('category').cat.codes  # Male -> 0, Female -> 1
df['ever_married'] = df['ever_married'].astype('category').cat.codes  # Yes -> 1, No -> 0

# One-hot encoding for non-binary categorical variables (work_type, residence_type)
df = pd.get_dummies(df, columns=['work_type', 'residence_type'], drop_first=True)

# Mapping smoking_status (optional: based on custom values)
smoking_map = {'never smoked': 0, 'formerly smoked': 1, 'smokes': 2}
df['smoking_status'] = df['smoking_status'].map(smoking_map)

# Check the updated dataframe
print(df.head())
```

```python
df = df.drop(columns=['id'])
```

```python

stroke_counts = df['stroke'].value_counts()

print(stroke_counts)
```

```python
from sklearn.utils import resample

# Separate the majority and minority classes
majority = df[df['stroke'] == 0]
minority = df[df['stroke'] == 1]

# Oversample the minority class to 2000 samples
minority_upsampled = resample(minority,
                              replace=True,
                              n_samples=2000,  # Exactly 2000 samples for minority class
                              random_state=42)

# Undersample the majority class to 2000 samples
majority_downsampled = resample(majority,
                                replace=False,
                                n_samples=2000,  # Exactly 2000 samples for majority class
                                random_state=42)

# Combine the upsampled minority class and the downsampled majority class
df_balanced = pd.concat([majority_downsampled, minority_upsampled])

# Shuffle the dataset
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Check the new class distribution
print(df_balanced['stroke'].value_counts())
```
