```python
import random
import matplotlib.pyplot as plt

def coinFlips(numFlips):
    heads = 0
    tails = 0
    probHistory = []

    for i in range(1, numFlips + 1):
        flip = random.choice(['Head', 'Tail'])
        if flip == 'Head':
            heads += 1
        else:
            tails += 1

        probHead = heads / i
        probTail = tails / i
        probHistory.append((i, probHead, probTail))

    return probHistory, heads, tails

# Scenario 1: 100 flips
prob100, heads100, tails100 = coinFlips(100)
# Scenario 2: 10,000 flips
prob10k, heads10k, tails10k = coinFlips(10000)

# Print final probabilities
print(f"(100 Flips: Heads = {heads100}, Tails = {tails100}")
print(f"P(Head) = {heads100 / 100:.2f}, P(Tail) = {tails100 / 100:.2f}")

print(f"\n10,000 Flips: Heads = {heads10k}, Tails = {tails10k}")
print(f"P(Head) = {heads10k / 10000:.4f}, P(Tail) = {tails10k / 10000:.4f}")

# Prepare data for plotting
x100 = [step for step, _, _ in prob100]
phead100 = [p for _, p, _ in prob100]

x10k = [step for step, _, _ in prob10k]
phead10k = [p for _, p, _ in prob10k]

# Plotting
# Plotting as Line Chart (No markers)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Line plot for 100 flips
plt.plot(x100, phead100, label='P(Head) - 100 Flips', color='blue')

# Line plot for 10,000 flips
plt.plot(x10k, phead10k, label='P(Head) - 10,000 Flips', color='green')

# Expected probability line
plt.axhline(y=0.5, color='gray', linestyle='--', label='Expected Probability (0.5)')

# Formatting
plt.title('Monte Carlo Simulation: Coin Flip Probabilities (Line Chart)')
plt.xlabel('Number of Flips')
plt.ylabel('P(Head)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Define the x range for the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)

# Plotting the standard normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')
plt.title('Standard Normal Distribution (μ=0, σ=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.axvline(0, color='black', linestyle='--', linewidth=1)
plt.legend()
plt.show()
```

```python
from scipy.stats import skew, kurtosis
np.random.seed(0)

# Normal distribution (baseline)
normal_data = np.random.normal(0, 1, 10000)

# Positive skew (exponential)
pos_skew_data = np.random.exponential(scale=1, size=10000)

# Negative skew (negated exponential)
neg_skew_data = -np.random.exponential(scale=1, size=10000)

# High kurtosis (leptokurtic) - using Laplace distribution
high_kurt_data = np.random.laplace(loc=0, scale=1, size=10000)

# Low kurtosis (platykurtic) - using uniform distribution
low_kurt_data = np.random.uniform(-np.sqrt(3), np.sqrt(3), 10000)

# Set up the plot
fig, axs = plt.subplots(3, 2, figsize=(14, 12))
axs = axs.ravel()
distributions = [
    ("Normal (μ=0, σ=1)", normal_data),
    ("Positive Skew", pos_skew_data),
    ("Negative Skew", neg_skew_data),
    ("High Kurtosis", high_kurt_data),
    ("Low Kurtosis", low_kurt_data)
]

# Plot each distribution
for i, (title, data) in enumerate(distributions):
    x_vals = np.linspace(min(data), max(data), 1000)
    y_vals = norm.pdf(x_vals, np.mean(data), np.std(data))
    axs[i].hist(data, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black')
    axs[i].plot(x_vals, y_vals, color='red')
    axs[i].set_title(f"{title}\nSkewness: {skew(data):.2f}, Kurtosis: {kurtosis(data):.2f}")
    axs[i].grid(True)

# Hide the unused subplot
axs[5].axis('off')
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import numpy as np

# Given values
sample_mean = 1.8
hypothesized_mean = 2
sample_std = 0.15
n = 10
alpha = 0.01

# Calculate t-statistic
t_stat = (sample_mean - hypothesized_mean) / (sample_std / np.sqrt(n))

# Degrees of freedom
df = n - 1

# Calculate p-value (left-tailed test)
p_value = stats.t.cdf(t_stat, df)

# Decision
reject_null = p_value < alpha

# Output
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")
print("Reject null hypothesis?" , "Yes" if reject_null else "No")
```

```python
import seaborn as sns
import scipy.stats as stats
import pandas as pd

# Load the built-in 'tips' dataset
df = sns.load_dataset("tips")
print(df.head())
# Separate tips by sex
male_tips = df[df["sex"] == "Male"]["tip"]
female_tips = df[df["sex"] == "Female"]["tip"]

# Perform Welch's t-test (independent samples, unequal variance)
t_stat, p_value = stats.ttest_ind(male_tips, female_tips, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Statistically significant difference in tip amounts between male and female customers.")
else:
    print(" No statistically significant difference in tip amounts between male and female customers.")
# Correlation Matrix - Select only numerical columns
numerical_columns = df.select_dtypes(include=["float64", "int64"])  # Select numerical columns only
correlation_matrix = numerical_columns.corr()  # Compute correlation matrix for numerical columns
print("\nCorrelation Matrix:\n", correlation_matrix)

# Heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()

# Scatterplot for 'total_bill' vs 'tip'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", alpha=0.6)
plt.title("Scatterplot: Total Bill vs Tip by gender")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()

# Scatterplot for 'size' vs 'tip'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="size", y="tip", hue="sex", alpha=0.6)
plt.title("Scatterplot: Size vs Tip by gender")
plt.xlabel("Size")
plt.ylabel("Tip")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# Load the 'tips' dataset
df = sns.load_dataset("tips")
print(df.head())

# Welch's t-test for tip difference by gender
male_tips = df[df["sex"] == "Male"]["tip"]
female_tips = df[df["sex"] == "Female"]["tip"]
t_stat, p_value = stats.ttest_ind(male_tips, female_tips, equal_var=False)
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Statistically significant difference in tip amounts between male and female customers.")
else:
    print("No statistically significant difference in tip amounts between male and female customers.")

# Compute correlation matrix of numerical features
numerical_df = df.select_dtypes(include=["float64", "int64"])
correlation_matrix = numerical_df.corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

# -----------------------------
# 💡 Python Equivalent of pairs.panels()
# -----------------------------

sns.set(style="ticks", color_codes=True)

# Define function to annotate correlation coefficients
def corrfunc(x, y, **kws):
    r, _ = stats.pearsonr(x, y)
    ax = plt.gca()
    ax.annotate(f"r = {r:.2f}", xy=(0.5, 0.5), xycoords=ax.transAxes,
                ha='center', va='center', fontsize=12, color='red')

# Create pairplot
pair_grid = sns.pairplot(numerical_df, diag_kind="kde", plot_kws={'alpha': 0.6})
pair_grid.map_upper(corrfunc)

# Title
plt.suptitle("Pairwise Correlations & Distributions (Tips Dataset)", y=1.02)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")
print(df.head())

# Select relevant variables
X = df[["total_bill"]]         # Independent variable
y = df["tip"]                  # Dependent variable

# Add constant term for intercept
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Display regression summary
print(model.summary())
# Scatterplot with regression line
sns.regplot(x="total_bill", y="tip", data=df, line_kws={"color": "red"})
plt.title("Linear Regression: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import statsmodels.formula.api as smf

# Load the dataset
df = sns.load_dataset("tips")

# Fit a multiple linear regression model
model = smf.ols(formula='tip ~ total_bill + size + sex + smoker + day + time', data=df).fit()

# Print the summary
print(model.summary())

import matplotlib.pyplot as plt
import seaborn as sns

# Get predictions
df['predicted_tip'] = model.fittedvalues
df['residuals'] = model.resid

# 1. Actual vs Predicted Tips
plt.figure(figsize=(8, 6))
sns.scatterplot(x='tip', y='predicted_tip', data=df, alpha=0.7)
plt.plot([df['tip'].min(), df['tip'].max()],
         [df['tip'].min(), df['tip'].max()],
         color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel("Actual Tip")
plt.ylabel("Predicted Tip")
plt.title("Actual vs Predicted Tips")
plt.legend()
plt.grid(True)
plt.show()

# 2. Residual Plot
plt.figure(figsize=(8, 6))
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, line_kws={"color": "red"})
plt.xlabel("Predicted Tip")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.grid(True)
plt.show()

```
