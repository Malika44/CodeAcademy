```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data Creation (you can replace this with a real dataset)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Region A Sales': [200, 220, 250, 270, 300, 310, 400, 420, 450, 470, 500, 550],
    'Region B Sales': [150, 180, 190, 200, 230, 250, 300, 320, 340, 360, 400, 430],
    'Region C Sales': [100, 120, 140, 150, 160, 180, 190, 210, 220, 230, 250, 280],
    'Advertising Budget': [50, 60, 55, 65, 70, 75, 85, 90, 95, 100, 110, 120]
}
df = pd.DataFrame(data)

# Display Data
print(df)

# Line Plot - Sales Trends Over Months
plt.figure(figsize=(5,5))
listofRegions = ['Region A Sales', 'Region B Sales', 'Region C Sales']
for region in listofRegions:
    plt.plot(df['Month'], df[region], label=region)
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Bar Chart - Total Sales by Region
total_sales = df[['Region A Sales', 'Region B Sales', 'Region C Sales']].sum()
total_sales.plot(kind='bar', figsize=(5,5), color=['blue', 'orange', 'green'])
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()

# Histogram - Distribution of Region A Sales
plt.figure(figsize=(5,5))
plt.hist(df['Region A Sales'], bins=5, color='blue', edgecolor='black')
plt.title('Distribution of Region A Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot - Advertising Budget vs. Region A Sales
plt.figure(figsize=(5,5))
plt.scatter(df['Advertising Budget'], df['Region A Sales'], color='purple', edgecolor='black')
plt.title('Advertising Budget vs. Region A Sales')
plt.xlabel('Advertising Budget')
plt.ylabel('Region A Sales')
plt.show()

# Pie Chart - Proportion of Sales by Region
total_sales.plot(kind='pie', figsize=(5,5), autopct='%1.1f%%', startangle=90, color=['blue', 'orange', 'green'])
plt.title('Proportion of Total Sales by Region')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = sns.load_dataset('iris')
```

```python
print(df.head())  # View the first few rows
print(df.info())  # View dataset information
```

```python
# Select numeric columns
df_numeric = df.select_dtypes(include=["number"])

# Compute the correlation matrix
correlation_matrix = df_numeric.corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap - Iris Dataset")
plt.show()
```
