```python
#1
ordersGreater = df[df['TotalAmount'] > 500]
print("Order Grater than 500:\n",ordersGreater)
electronicsOrders = df[df['Category'] == 'Electronics']
print(f"orders in the Electronics category:\n {electronicsOrders}")
```

```python
#2
print("orders where the Category is Clothing and the Status is Completed \n", df[(df['Category']=='Clothing') & (df['Status']== 'Completed')])
print( "orders where the Quantity is greater than 5 or the Status is Pending:\n",df[(df['Quantity'] > 5) | (df['Status'] == 'Pending')])
```

```python
#3
print("orders where the TotalAmount is between 200 and 1000:\n", df[(df['TotalAmount'] >= 200) & (df['TotalAmount'] <= 1000)])
```

```python
#4
print("List all orders where the CustomerName starts with 'A':\n",df[df['CustomerName'].str.startswith('A',na=False)])
print("Find all orders where the Category contains 'Home':\n",df[df['Category'].str.contains('Home',na=False)])
```

```python
#5
print("orders where the CustomerName is missing (NaN):\n",df[df['CustomerName'].isna()])
print("orders where the OrderDate is not missing:\n",df[df['OrderDate'].notna()])
```

```python
#filtering
#1
print('rows where the Status is "Completed:\n"',df[df['Status'] == 'Completed'])
```

```python
#2
print("DataFrame with only the OrderID , CustomerName , and TotalAmount columns:\n", df[['OrderID', 'CustomerName', 'TotalAmount']])
```

```python
#3
mask = df['Price'] > 100
priceOver100 = df[mask]
print(priceOver100)
```

```python
#4
x= df.loc[df['TotalAmount'] > 500, ['OrderID', 'Status']]
print(x)
x.to_csv("high_value_orders.csv", index =True)
```
