```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
import statistics
import time

# Set up the driver
driver = webdriver.Chrome()  # Or use the path: webdriver.Chrome(executable_path="path_to_driver")
driver.get("https://books.toscrape.com/")
time.sleep(2)  # Optional delay to let the page load

# Locate all book elements on the page
books = driver.find_elements(By.CLASS_NAME, "product_pod")

# Data lists
book_data = []

for book in books:
    title = book.find_element(By.XPATH, ".//h3/a").get_attribute("title")
    price = book.find_element(By.CLASS_NAME, "price_color").text.strip('£')
    availability = book.find_element(By.CLASS_NAME, "availability").text.strip()
    star_class = book.find_element(By.CLASS_NAME, "star-rating").get_attribute("class").split()[-1]

    # Convert star class to numeric value
    star_dict = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
    star_rating = star_dict.get(star_class, 0)

    book_data.append([title, float(price), availability, star_rating])

# Save to CSV
with open('books.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Price', 'Availability', 'Star Rating'])
    writer.writerows(book_data)

# Summary statistics
total_books = len(book_data)
average_price = round(statistics.mean([item[1] for item in book_data]), 2)

print(f"Total number of books: {total_books}")
print(f"Average price: £{average_price}")

# Close the driver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import json
import time

# Set up WebDriver
driver = webdriver.Chrome()
driver.get("https://www.bbc.com/news/technology")
time.sleep(3)  # Wait for the page to load

# Find all article blocks
articles = driver.find_elements(By.XPATH, '//div[contains(@class, "gs-c-promo")]')

news_data = []

for article in articles:
    try:
        # Headline
        headline_elem = article.find_element(By.XPATH, './/h3')
        headline = headline_elem.text.strip()

        # URL
        link_elem = article.find_element(By.XPATH, './/a[@href]')
        relative_url = link_elem.get_attribute("href")
        url = relative_url if relative_url.startswith("http") else f"https://www.bbc.com{relative_url}"

        # Summary (if available)
        try:
            summary_elem = article.find_element(By.XPATH, './/p')
            summary = summary_elem.text.strip()
        except:
            summary = ""

        news_data.append({
            "Headline": headline,
            "URL": url,
            "Summary": summary
        })
    except:
        continue  # Skip if essential elements are missing

# Save data to JSON
with open("bbc_technology_news.json", "w", encoding='utf-8') as f:
    json.dump(news_data, f, indent=4, ensure_ascii=False)

# Output to terminal
print(json.dumps(news_data, indent=2, ensure_ascii=False))

# Close the driver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import json

options = webdriver.ChromeOptions()
# options.add_argument("--headless")  # Optional headless mode
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(options=options)
driver.get("https://www.bbc.com/news/technology?set-language=en&set-location=GB")


print("Final URL:", driver.current_url)

# Optional: Handle cookie popup
try:
    consent = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Yes, I agree")]'))
    )
    consent.click()
    print("✔ Cookie consent handled.")
except:
    print("❌ No consent popup detected.")

# Screenshot to debug layout
driver.save_screenshot("bbc_tech_page.png")

# Wait for articles
try:
    headline_blocks = WebDriverWait(driver, 20).until(
        EC.visibility_of_all_elements_located(
            (By.XPATH, '//a[contains(@class, "gs-c-promo-heading")]')
        )
    )

    news_data = []

    for block in headline_blocks:
        try:
            headline = block.text.strip()
            url = block.get_attribute("href")
            summary = "No summary available"
            try:
                parent = block.find_element(By.XPATH, "..")
                summary_elem = parent.find_element(By.XPATH, './/p')
                summary = summary_elem.text.strip()
            except:
                pass

            if headline and url:
                news_data.append({
                    "Headline": headline,
                    "URL": url,
                    "Summary": summary
                })
        except Exception as e:
            print("Skipping one article:", e)

    with open("bbc_technology_news.json", "w", encoding='utf-8') as f:
        json.dump(news_data, f, indent=4, ensure_ascii=False)

    print(f"✅ Extracted {len(news_data)} articles.")
except Exception as e:
    print("❌ Failed to find headlines:", e)

driver.quit()
```
